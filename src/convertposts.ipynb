```python
import frontmatter
import os
import glob
from datetime import datetime
```

```python
input_dir = 'toconvert'
output_dir = 'drafts'
```

```python
def convert_date_format(datestr, oldformat="%Y-%m-%d %I:%M %p", newformat="%Y-%m-%d %H:%M:00"):
    dtobj = datetime.strptime(datestr, oldformat)
    return dtobj.strftime(newformat)

def convert_file(filepath, output_dir=output_dir):
    print(f'reading {filepath}')
    filename = filepath.split('/')[1]

    with open(filepath, 'r') as file:
        post = frontmatter.load(file)
    
    # convert the date
    newdatestr = convert_date_format(post.metadata['date'])

    # create and write new file
    newfile = os.path.join(output_dir, filename)
    newfilestr = f'''---
date: {newdatestr}
---

{post.content}
'''
    print(f'writing {newfile}')
    with open(newfile, 'w') as file:
        file.write(newfilestr)

    return 0
```

```python
# I used to use os.listdir, but this might work better
# https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory#3207973

postfiles = glob.glob(f'{input_dir}/*.md')

for post in postfiles:
    convert_file(post)
```
